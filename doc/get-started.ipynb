# 快速上手

先决条件：您应该对 [Sphinx 生态系统](http://www.sphinx-doc.org/) 有所了解，并且已经在本地计算机上安装了 Sphinx。

## 安装并激活主题

首先使用 `pip` 安装 `xyzstyle`：

```shell
pip install xyzstyle
```

然后，在您的 Sphinx 配置文件（ `conf.py` ）中激活主题：
```{code-block} python
:caption: conf.py

...
html_theme = "xyzstyle"
...
```

这将为您的文档激活 `xyzstyle` 主题。


## 进阶教程

```{gallery-grid}
:grid-columns: 1 2 2 3

- header: "{fab}`config;pst-color-primary` Sphinx 基础配置"
  content: "[](./configs/base) 包含了可用的基本配置选项说明，可以直接使用。"
- header: "{fas}`book;pst-color-primary` sphinx-book-theme 中文文档"
  content: "`syzstyle` 的父主题是 `sphinx-book-theme` 主题，可以查阅[中文文档](https://daobook.github.io/sphinx-book-theme/)。"
- header: "{fas}`book;pst-color-primary` pydata-sphinx-theme 中文文档"
  content: "`sphinx-book-theme` 的父主题是 `pydata-sphinx-theme` 主题，可以查阅[中文文档](https://daobook.github.io/pydata-sphinx-theme/)。"
- header: "{fas}`book;pst-color-primary` MyST-NB 中文文档"
  content: "可以查阅[MyST-NB 中文文档](https://daobook.github.io/MyST-NB/)。"
- header: "{fas}`book;pst-color-primary` MyST-Parser 中文文档"
  content: "可以查阅[MyST-Parser 中文文档](https://daobook.github.io/MyST-Parser/)。"
```
